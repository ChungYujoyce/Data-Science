```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('processedDatano.csv')
label = df['RainToday']
df = df.drop(['RainToday','w'], axis=1)
df
```

```python
from sklearn.preprocessing import normalize

df = normalize(df, norm='l2', axis=1)
df = pd.DataFrame(df)
df.head()
```

```python
df.columns = ['MinTemp', 'MaxTemp', 'Evaporation', 'Sunshine', 'WindGustSpeed',
       'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 'Humidity3pm',
       'Pressure9am', 'Pressure3pm', 'Cloud9am', 'Cloud3pm', 'Temp9am',
       'Temp3pm', 'RISK_MM']
df['RainToday'] = label
df.head(15)
```

```python
df.to_csv('./processedDataNorm.csv', index = False)
```
