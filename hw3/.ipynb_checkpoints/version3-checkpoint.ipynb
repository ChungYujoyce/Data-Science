```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('processedDataNorm.csv')
train_end_idx = 35156 # training.csv size

X_ans = df.drop(columns = ['RainToday']).values[train_end_idx:, :]
X_ans = pd.DataFrame(X_ans)
X_ans.columns = ['MinTemp', 'MaxTemp', 'Evaporation', 'Sunshine', 'WindGustSpeed', 'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 'Humidity3pm', 'Pressure9am', 'Pressure3pm', 'Cloud9am', 'Cloud3pm', 'Temp9am', 'Temp3pm', 'RISK_MM'] 
```

```python
X = df.drop('RainToday',axis=1).values[:train_end_idx, :]
y = df['RainToday'].values[:train_end_idx]

X = pd.DataFrame(X)
y = pd.DataFrame(y)
X.columns = ['MinTemp', 'MaxTemp', 'Evaporation', 'Sunshine', 'WindGustSpeed', 'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 'Humidity3pm', 'Pressure9am', 'Pressure3pm', 'Cloud9am', 'Cloud3pm', 'Temp9am', 'Temp3pm', 'RISK_MM'] 
y.columns = ['RainToday']
X.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.utils import resample
#split data into test and training sets

X_train1, X_test1, y_train1, y_test1 = train_test_split(X, y, test_size=0.3, random_state=42)
#combine them back for resampling
train_data = pd.concat([X_train1, y_train1], axis=1)
# separate minority and majority classes
negative = train_data[train_data.RainToday==0]
positive = train_data[train_data.RainToday==1]
# upsample minority
pos_upsampled = resample(positive,
 replace=True, # sample with replacement
 n_samples=len(negative), # match number in majority class
 random_state=101) # reproducible results
# combine majority and upsampled minority
upsampled = pd.concat([negative, pos_upsampled])
# check new class counts
upsampled.RainToday.value_counts()
```

```python
X = upsampled.drop('RainToday', axis=1)
y = upsampled['RainToday']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
len(X_train)
```
